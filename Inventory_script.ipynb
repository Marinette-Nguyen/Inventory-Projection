# Defined func

### read_html

```python


def concat_files_html(folder_path, pattern,row):
    files = [f for f in os.listdir(folder_path) if pattern in f]
    df_combine = pd.DataFrame()
    df_combine = df_combine.iloc[0:0]
    for filename in files:
        file_path = os.path.join(folder_path, filename)
        print(f"Processing file: {file_path}")
        
        try:
            file = pd.read_html(file_path, header= row)
            df = file[0]
            df["Source"] = filename
            df_combine = pd.concat([df_combine, df], ignore_index=True)
            print(f"Processed file: {filename}")
        except Exception as e:
            print(f"Failed to process file {filename}: {e}")
    
    return df_combine

```

### read_excel

```python
def concat_files_excel(folder_path, pattern,row, cols):
    files = [f for f in os.listdir(folder_path) if pattern in f]
    df_combine = pd.DataFrame()
    for filename in files:
        file_path = os.path.join(folder_path, filename)
        print(f"Processing file: {file_path}")
        
        try:
            file = pd.read_excel(file_path, header=row, usecols =cols)
            file["Source"] = filename
            df_combine = pd.concat([df_combine, file], ignore_index=True)
            print(f"Processed file: {filename}")
        except Exception as e:
            print(f"Failed to process file {filename}: {e}")
    
    return df_combine
```

### func define

```python

def allocate_strict(group, qty_bom_col, qty_col, alloc_col):
    bom_qty = group[qty_bom_col].iloc[0]
    cum_demand = 0
    allocated = []
    allocated_done = False

    for qty in group[qty_col]:
        if allocated_done:
            allocated.append(qty)
        else:
            cum_demand += qty
            if allocated_done:
                allocated.append(0)
            elif cum_demand < bom_qty:
                allocated.append(0)
            else:
                alloc = min(qty, cum_demand - bom_qty)
                allocated.append(alloc)
                allocated_done = True

    group[alloc_col] = allocated
    return group
```
